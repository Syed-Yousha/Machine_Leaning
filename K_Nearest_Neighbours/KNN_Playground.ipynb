```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline 
```

```python
df = pd.read_csv('Classified Data', index_col = 0)
```

```python
df.head()
```

```python
#Our data is unscaled and can cause trouble for our model so we have to scale it first
from sklearn.preprocessing import StandardScaler
```

```python
scalar =  StandardScaler()
scalar.fit(df.drop('TARGET CLASS', axis=1))
```

```python
scaled_features = scalar.transform(df.drop('TARGET CLASS', axis=1))
```

```python
scaled_features 
```

```python
df_feat = pd.DataFrame(scaled_features, columns = df.columns[:-1])
```

```python
df_feat.head() #Standard Scaled version of dataframe without target class
```

```python
#Lets split the data
from sklearn.model_selection import train_test_split
```

```python
X = df_feat
y = df['TARGET CLASS']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#Using KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
#We can make more accurate Prediction by Changing the K value
```

```python
error_rate = []

for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 40), error_rate, color='blue', linestyle='dashed',
         marker='o', markerfacecolor='red', markersize=10)
plt.title('Error rate vs K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
# Error is lowest when K is 35 
```

```python
#lets choose 17 for now because graph is too bumpy
knn = KNeighborsClassifier(n_neighbors=17)
knn.fit(X_train, y_train)
predictions = knn.predict(X_test)

print(classification_report(y_test, predictions))
print("\n\n")
print(confusion_matrix(y_test, predictions))
```
